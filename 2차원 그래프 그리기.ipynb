```python
#3차 함수 f(x) = (x-2)x(x+2)

#리스트 2-(1)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def f(x):
    return (x-2) * x * (x+2)

#리스트2-(2)
print(f(1))

#리스트2-(3)
print(f(np.array([1, 2, 3])))

#리스트 2-(4)
x = np.arange(-3, 3.5, 0.5)
print(x)

#리스트 2-(5)  2-(4) 보다 설정하기 쉬운 linspace(n1, n2, n)를 사용n1~n2까지 일정간격 n개의 구간으로 나누 값을 리턴
x = np.linspace(-3, 3, 10)
print(np.round(x, 2))
#linspace를 사용하면 n2가 x의 그래프에 포함되고 그래프의 세밀함도 n을 수정하면 간단하게 제어 할 수 있다.

#리스트 2-(6)
plt.plot(x, f(x))
plt.show()

```

```python
#리스트 2-(7) 그래프 장식 하기
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#함수를 정의
def f2(x, w):
    return (x-w)*x*(x+2) # (A)함수 정의

#x를 정의
x = np.linspace(-3, 3, 100) #(B) x를 100분할 하기

#차트 묘사
plt.plot(x, f2(x, 2), color='black', label='$w=2$') #(C)
plt.plot(x, f2(x, 1), color='cornflowerblue', label='$w=1$') #(D)
plt.legend(loc="upper left") #(E)범례 표시
plt.ylim(-15, 30) #(F) y축의 범위
plt.xlim(-4, 4) # x축 범위를 지정할 수 있다.
plt.title('$f_2(x)$') #(G) 제목
plt.xlabel('$x$') #(H) x 라벨
plt.ylabel('$y$') #(I) y 라벨
plt.grid(True) #(J) 그리드
plt.show()
```

```python
#리스트 2-(9)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#함수를 정의
def f2(x, w):
    return (x-w)*x*(x+2) # (A)함수 정의

plt.figure(figsize=(10, 3)) #(A) figure 지정
plt.subplots_adjust(wspace=0.5, hspace=0.5) # (B) 그래프의 간격을 지정
for i in range(6):
    plt.subplot(2, 3, i+1) #(C) 그래프 묘사의 위치를 지정
    plt.title(i+1)
    plt.plot(x, f2(x, i), 'k')
    plt.ylim(-20, 20)
    plt.grid(True)
plt.show()
```
