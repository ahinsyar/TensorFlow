```python
# 3-(1)
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# f3 
def f3(x0, x1):
    r = 2 * x0**2 + x1**2
    ans=r*np.exp(-r)
    return ans

# x0, x1  f3 
xn = 50
x0 = np.linspace(-2, 2, xn) # (A)
x1 = np.linspace(-2, 2, xn) # (B)
y = np.zeros((len(x0), len(x1))) # (C)
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1]) # (D)
        
# 3-(2)
print(x0)

# 3-(3)
print(np.round(y, 1))

# 3-(4)
plt.figure(figsize=(3.5, 3))
plt.gray() #(A)
plt.pcolor(y) #(B)
#plt.contour(y)
plt.colorbar() #(C)
plt.show()

# 3-(5)
#import importlib
#importlib.import_module('mpl_toolkits.mplot3d').Axes3D
from mpl_toolkits.mplot3d import Axes3D #(A)

xx0, xx1 = np.meshgrid(x0, x1) #(B)

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(1, 1, 1, projection='3d') #(C)
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=0.3, color='blue', edgecolor='black')#(D)
ax.set_zticks((0, 0.2)) #(E)
ax.view_init(45, -120)
plt.show()

# 3-(6)
print(x0)
print(x1)

# 3-(7)
print(xx0)

# 3-(8)
print(xx1)

# 3-(9)  (contour)
xn = 50
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)

y = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])
        
xx0, xx1 = np.meshgrid(x0, x1) #(A)

plt.figure(1, figsize=(4, 4))
cont = plt.contour(xx0, xx1, y, 5, colors='black') #(B)
cont.clabel(fmt='%3.2f', fontsize=8)#(C)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.show()
```
